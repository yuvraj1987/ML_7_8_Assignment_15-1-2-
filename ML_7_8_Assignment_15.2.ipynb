```python
## In this Assignment student have to find the the frequency of words in a webpage 
## User can use urllib and Beautiful Soup to extract text from webpage
```

```python
Given_URL = "http://php.net/"
from bs4 import  BeautifulSoup
import urllib.request
import nltk
```

```python
response = urllib.request.urlopen(Given_URL)
html     = response.read()
soup = BeautifulSoup(html,"html5lib")
```

```python
soup
```

```python
Text = soup.get_text()
```

```python
import re
tokens = re.findall('\w+',Text)
tokens[:10]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set_style('darkgrid')

nlp_words = nltk.FreqDist(tokens)
nlp_words.plot(20)
```

```python
## 
from collections import defaultdict
Freq_Dict = defaultdict(int)
for k in tokens:
    Freq_Dict[k] +=1 
```

```python
import operator
sorted_x = sorted(Freq_Dict.items(), key=operator.itemgetter(1))
```

```python
sorted_x.reverse()
```

```python
sorted_x
```

```python
import pandas as pd
Word_Df = pd.DataFrame(sorted_x, columns=["Words","Frequecy"])
```

```python
Word_Df.head()
```

```python
Word_Df.head(10).plot(kind='bar',x='Words', y="Frequecy",figsize=(10,5),style =list,fontsize = 20,
                     position =0.8,title ="Top Ten Words")
```

```python
Word_Df.head(10).plot(kind='barh',x='Words', y="Frequecy",figsize=(10,5),style =list,fontsize = 20,
                     position =0.8,title ="Top Ten Words")
```
